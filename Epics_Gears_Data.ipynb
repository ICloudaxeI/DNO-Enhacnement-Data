Create a function that calls the name of the item and getting their row number to convert in to a string

```python
def get_item_info(keyword):
    # Filter the DataFrame to get all columns for items containing the keyword
    item_info = df[df['Item Name'].str.contains(keyword, case=False)]

    return item_info

additional_words = ['Destroyer', 'Illumia',             # L grade warrior
                    'Troublemaker ', 'Tracker',         # L grade archer
                    'Saviour', 'Guardian Knight',       # L grade cleric
                    'Contractor', 'Terminator',         # L grade sorceress
                    'Changer', 'Coordinator',           # L grade academic
                    'Soul Handler', 'Dancer',           # L grade kali
                    'Slaughter', 'Meditation',          # L grade assassin
                    'Victoria']                         # L grade lancea

def filter_items(item_info, additional_words):
    # Convert additional words list to lowercase
    additional_words = [word.lower() for word in additional_words]

    # Function to check if item name contains any additional words
    def contains_additional_words(item_name):
        item_name_lower = item_name.lower()
        for word in additional_words:
            if word in item_name_lower:
                return True
        return False

    # Apply the filter
    filtered_info = item_info[~item_info['Item Name'].apply(contains_additional_words)]
    return filtered_info
```

# Making the bulk csv file of Unique set in one csv

Create a function that combines all `str_row_indices` into an array and convert it into csv

```python
# List containing all item_info_
all_item_info_list = [
    item_info_E1, item_info_E2,
    item_info_E3, item_info_E4,
    item_info_E5, item_info_E6,
    item_info_E7, item_info_E8,
    item_info_E9, item_info_E10,
    item_info_E11, item_info_E12,
    item_info_E13
]

# Output CSV file path in Google Drive
output_csv_file = 'all_E_gears.csv'

# Write the filtered rows to the CSV file in Google Drive
with open('/content/drive/MyDrive/Enhancement_data/Epics/' + output_csv_file, 'w', newline='') as output_csv:
    writer = csv.writer(output_csv)
    # Write Header
    writer.writerow(['Item Name', 'EnchantLevel', 'EnchantRatio', 'BreakRatio', 'MinDowngrade', 'MaxDowngrade', 'Gold'])

    # Write rows corresponding to the filtered DataFrames
    for item_info in all_item_info_list:
        # Write each row to the output CSV file
        for _, row in item_info.iterrows():
            writer.writerow(row)

print(f"Combined rows saved to '{output_csv_file}' in Google Drive")
```
